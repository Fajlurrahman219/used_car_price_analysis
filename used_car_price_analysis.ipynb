```python
df1=pd.get_dummies(df,columns=['Fuel_Type','Seller_Type','Transmission'],drop_first=True)
df1
```

```python
df2=df.groupby(['Car_Name'])['Selling_Price'].mean().reset_index().rename(columns={'Selling_Price':'Car_Name_Mean_Selling_Price'})
df2
```

```python
df3=pd.merge(df1,df2,how='inner',on='Car_Name')
df3
```

```python
df3.drop(columns=['Car_Name'],inplace=True)
df3
```

```python
x=df3.drop(columns=['Selling_Price'])
y=df3['Selling_Price']
```

```python
from sklearn.preprocessing import StandardScaler

st=StandardScaler()
x_scaled=st.fit_transform(x)
x_scaled
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=1)

print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
x_scaled.shape
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
from sklearn.metrics import r2_score

y_pred_train=lr.predict(x_train)
y_pred_test=lr.predict(x_test)
```

```python
r2_score(y_train,y_pred_train)
```

```python
r2_score(y_test,y_pred_test)
```
